```python
# TAKE THE TABULAR INFO FROM THE SHAPEFILE AND SAVE AS A DF FOR ANALYSIS

import geopandas as gpd
import pandas as pd

# Read the shapefile
classified_shp = gpd.read_file("Classified_Parcels.shp")

# Extract tabular data (excluding geometry) and convert it to a DataFrame
classified_df = classified_shp.drop(columns='geometry')

# Save and preview
classified_df.to_csv("classified_df.csv", index=False)
classified_df.head()
```

```python
print(classified_df.dtypes)
```

```python
# RENAME COLUMNS AND CONVERT DATA TYPES TO ORIGINALS

# Rename multiple columns
column_mapping = {'PTY_TYP': 'PTY.TYPE', 
                  'X_CENTR': 'X.CENTROID', 
                  'Y_CENTR': 'Y.CENTROID',
                  'NUM_UNI': 'NUM.UNITS',
                  'WOOD_DE': 'WOOD.DECK.SQFT',
                  'WOOD_ST': 'WOOD.STOVE.COUNT',
                  'ROOF_CO': 'ROOF.COMBUST',
                  'SIDING_': 'SIDING.COMBUST',
                  'DETACHE': 'DETACHED.STORAGE',
                  'GISAcrs': 'GIS.ACRES',
                  'WUI_CLASS': 'WUI.CLASS',
                  'RISK_SCORE': 'RISK.SCORE'
                 }

classified_df = classified_df.rename(columns = column_mapping)

# Convert the data types
classified_df['NUM.UNITS'] = classified_df['NUM.UNITS'].fillna(0).astype(int)
classified_df['WOOD.STOVE.COUNT'] = classified_df['WOOD.STOVE.COUNT'].fillna(0).astype(int)
classified_df['ROOF.COMBUST'] = pd.to_numeric(classified_df['ROOF.COMBUST'], errors='coerce').astype('Int64')
classified_df['SIDING.COMBUST'] = pd.to_numeric(classified_df['SIDING.COMBUST'], errors='coerce').astype('Int64')
classified_df['DETACHED.STORAGE'] = pd.to_numeric(classified_df['DETACHED.STORAGE'], errors='coerce').astype('Int64')
classified_df['WUI.CLASS'] = classified_df['WUI.CLASS'].astype(int)
classified_df['RISK.SCORE'] = classified_df['RISK.SCORE'].astype(int)

# Convert the two categorical columns to categorical
classified_df['WUI.CLASS'] = pd.Categorical(classified_df['WUI.CLASS'],
                                             categories = [1,2,3,4,5,6,7,8],
                                             ordered=False
                                            )
classified_df['RISK.SCORE'] = pd.Categorical(classified_df['RISK.SCORE'],
                                              categories=[1, 2, 3, 4, 5],
                                              ordered=True
                                             )

print(classified_df.dtypes)
classified_df.head()
```

```python
# CREATE A TRAIN-TEST SPLIT AND FIT A RF CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'df' is your DataFrame
# Replace 'your_predictor_columns' with the actual column names you want to use as predictors

# Select predictors and target variable
predictors = df[['predictor_column1', 'predictor_column2', 'predictor_column3']]
target = df['RISK.SCORE']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=42)

# Create a random forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model on the training data
rf_classifier.fit(X_train, y_train)

# Make predictions on the test data
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Print classification report
print(classification_report(y_test, y_pred))
```
